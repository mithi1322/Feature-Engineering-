Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.

```python
import numpy as np
# clean the dataset
df=df.drop(df.index[10472])
df['Size']=df['Size'].str.replace('M','000')
df['Size']=df['Size'].str.replace('k','')
df['Size']=df['Size'].replace('Varies with device',np.nan)
df=df.drop_duplicates(subset=['App'],keep='first')
df['Size']=df['Size'].astype(float)
# Create a scatter plot
plt.scatter(df['Size'], df['Rating'], alpha=0.4)
plt.title('Size vs Rating')
plt.xlabel('Size (KB)')
plt.ylabel('Rating')
plt.show()
```

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python

# Group the data by Type and calculate the mean price for each group
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df[cols]=df[cols].str.replace(item,'')
df['Price']=df['Price'].astype('float')
avg_prices = df.groupby('Type')['Price']
# Create a bar chart to compare the average prices by app type
plt.bar(avg_prices.index, avg_prices.values)
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Price ($)')
plt.show()
```
